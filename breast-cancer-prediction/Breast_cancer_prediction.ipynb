<h1>Featuure Extraction and Engineering based on EDA</h1>

```python
#Dropping less relevent features

df.drop('id', axis=1, inplace=True)

corr_matrix = df.corr().abs()
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
tri_df = corr_matrix.mask(mask)
tri_df = tri_df['diagnosis']
to_drop = [x for x in tri_df.index if tri_df[x]<0.1]
print(to_drop)
df = df.drop(to_drop, axis=1)
```

```python
#Feature Engineering

df['shape_mean_mean'] = (df['radius_mean'] + df['perimeter_mean'] + df['area_mean']) / 3
df['shape_se_mean'] = (df['radius_se'] + df['perimeter_se'] + df['area_se']) / 3
df['shape_worst_mean'] = (df['radius_worst'] + df['perimeter_worst'] + df['area_worst']) / 3

df = df.drop(['radius_mean', 'perimeter_mean', 'area_mean', 'radius_se', 'perimeter_se', 'area_se', 'radius_worst', 'perimeter_worst', 'area_worst'], axis=1)
```

<h1>Training the SDVM model for tumor classification</h1>

```python
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train ,y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape, X_test.shape)
```

```python
from sklearn.svm import SVC
model = SVC(C=15, gamma=0.01, probability=True)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

<h1>Evaluating the model's performance</h1>

```python
#Confusion matrix and heatmap

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
model_acc = accuracy_score(y_test, y_pred)
print(f"Accuracy: {model_acc}\n")
print("Heatmap:")
sns.heatmap(confusion_matrix(y_test, y_pred), linewidths=1, annot=True, fmt = ".2f")
plt.show()
print(f"Classification report:\n\n{classification_report(y_test, y_pred)}")
```

```python
#Plotting the learning curve

from sklearn.model_selection import learning_curve

train_sizes, train_scores, val_scores = learning_curve(model, X_train, y_train, cv=5)
train_scores_mean = np.mean(train_scores, axis=1)
val_scores_mean = np.mean(val_scores, axis=1)

plt.plot(train_sizes, train_scores_mean, 'o-', label='Training Score')
plt.plot(train_sizes, val_scores_mean, 'o-', label='Validation Score')
plt.xlabel('Number of Training Samples')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend()
plt.show()
```
